```python
with open('preed_for_cats.txt') as f:
  lines = f.readlines()
d = {}
categories = ['Food', 'Interior', 'Price', 'Service', 'Whole']
for line in lines:
  id = line.split('\t')[0]
  d[id] = {}
for line in lines:
  id = line.split('\t')[0]
  cat = line.split('\t')[1]
  ton = line.split('\t')[-1].strip('\n')
  if cat not in d[id].keys():
    d[id][cat] = []
  d[id][cat].append(ton)
```

```python
res = {}
for id, vs in d.items():
  res[id] = {}
  for c in categories:
    res[id][c] = 'absence'
  for cat, tons in vs.items():
    if 'positive' in tons and 'negative' in tons:
      res[id][cat] = 'both'
    elif 'positive' in tons:
      res[id][cat] = 'positive'
    else:
      res[id][cat] = 'negative'
res
```

```python
with open('pred_cats.txt', 'w') as f:
  for id, v in res.items():
    for cat, ton in v.items():
      f.write(f'{id}\t{cat}\t{ton}\n')
```
